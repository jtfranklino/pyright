```python
import subprocess

try:
    # Find the PID listening on port 11434
    pid_command = "lsof -t -i:11434"
    pid_output = subprocess.check_output(pid_command, shell=True).decode('utf-8').strip()

    if pid_output:
        pid = int(pid_output)
        print(f"Found process with PID {pid} listening on port 11434. Killing process...")
        # Kill the process
        kill_command = f"kill -9 {pid}"
        subprocess.run(kill_command, shell=True, check=True)
        print(f"Process {pid} killed successfully.")
    else:
        print("No process found listening on port 11434.")
except subprocess.CalledProcessError as e:
    print(f"Error: {e}")
    print(f"Stderr: {e.stderr}")
except FileNotFoundError:
    print("Error: lsof command not found. Please ensure lsof is installed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import requests
from pathlib import Path
import subprocess
import time

MODEL = "llama3.1"  # or llama3.2:1b if your machine is slow

def ensure_ollama_server_running(max_retries=10, retry_delay=5):
    for attempt in range(max_retries):
        try:
            # Try to ping the Ollama API
            response = requests.get("http://127.0.0.1:11434/api/tags", timeout=1)
            if response.status_code == 200:
                print(f"Ollama server is running and responsive on attempt {attempt + 1}.")
                return True
        except requests.exceptions.ConnectionError:
            print(f"Ollama server not yet responsive on attempt {attempt + 1}. Retrying in {retry_delay} seconds...")
        except Exception as e:
            print(f"An unexpected error occurred while checking Ollama server: {e}")

        if attempt < max_retries - 1:
            # If not responsive, try to ensure it's started
            try:
                pid_command = "lsof -t -i:11434"
                pid_output = subprocess.check_output(pid_command, shell=True).decode('utf-8').strip()
                if not pid_output:
                    print("No Ollama process found, starting server...")
                    get_ipython().system_raw('ollama serve > /tmp/ollama.log 2>&1 &')
                    time.sleep(retry_delay) # Wait for it to potentially start
            except subprocess.CalledProcessError:
                print("Error checking for Ollama process, attempting to start server...")
                get_ipython().system_raw('ollama serve > /tmp/ollama.log 2>&1 &')
                time.sleep(retry_delay)
            time.sleep(retry_delay)

    print("Failed to start or connect to Ollama server after multiple attempts.")
    return False

def ask_ollama(prompt, model=MODEL, temperature=0.1):
    r = requests.post(
        "http://127.0.0.1:11434/api/generate",
        json={
            "model": model,
            "prompt": prompt,
            "stream": False,
            "options": {"temperature": temperature}
        },
        timeout=600
    )
    r.raise_for_status()
    return r.json()["response"]

def run_skill(skill_name, task):
    skill_path = Path(f"ECC/.agents/skills/{skill_name}/SKILL.md")
    skill = skill_path.read_text(errors="ignore")
    prompt = f"""
You are operating under this ECC skill:

{skill}

Task:
{task}

Rules:
- Do not invent facts.
- If a document does not support a conclusion, say NEEDS VERIFICATION.
- Give actionable output.
"""
    return ask_ollama(prompt)

# Ensure Ollama server is running before making the request
if ensure_ollama_server_running():
    print(run_skill(
        "agent-introspection-debugging",
        "Debug why my Ollama forensic document pipeline may fail on scanned PDFs."
    ))
else:
    print("Ollama server is not running. Cannot execute skill.")
```
